```python
from ema_workbench import load_results
from ema_workbench.analysis import feature_scoring
import matplotlib.pyplot as plt
import seaborn as sns
from model.misinfo_model import MisinfoPy
from dmdu.utils_dmdu import (get_levers, get_uncertainties)
import os
```

```python
# Levers
target_dir = os.path.join(os.getcwd(), 'data', 'open_exploration_50_policies_SIT')
results = load_results(target_dir)
exp, out = results

levers = [l.name for l in get_levers()]
exp_only_uncert = exp.drop(columns=levers + ["policy"])

sns.set_style("whitegrid")
fs = feature_scoring.get_feature_scores_all(exp_only_uncert, out)
sns.set(rc={'figure.figsize': (20, 15)})
sns.heatmap(fs, cmap="viridis", annot=True)
plt.show()
```
